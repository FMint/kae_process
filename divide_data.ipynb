```python
#生成all.txt
import os

data_dir = r"F:\fsy\project\kae_process\kae_smpl\sad"
txt_dir = r"F:\fsy\project\kae_process\infor\sad"

# 确保目录存在
os.makedirs(txt_dir, exist_ok=True)

all_path = os.path.join(txt_dir, "all.txt")
with open(all_path, "w", encoding="utf-8") as all_f:
    for f in os.listdir(data_dir):
        if f.endswith(".npz"):
            # 去掉 .npz
            name = f[:-4]
            all_f.write(name + "\n")
```

```python
#划分数据集 0.8 0.15 0.05 train test train_val
all_f = open(txt_dir+r"\all.txt","r")
lines = all_f.readlines()
num_lines = len(lines)
train_end = int(num_lines * 0.8)
test_end = int(num_lines * 0.95)
train_lines = lines[:train_end]
test_lines = lines[train_end:test_end]
val_lines = lines[test_end:]
with open(txt_dir+r"\train.txt","w") as f:
    f.writelines(train_lines)
with open(txt_dir+r"\test.txt","w") as f:
    f.writelines(test_lines)
with open(txt_dir+r"\val.txt","w") as f:
    f.writelines(val_lines)
with open(txt_dir+r"\train_val.txt","w") as f:
    f.writelines(train_lines + val_lines)
```

```python
# 合并不同情绪的数据集文件
import os
from pathlib import Path

base_dir = Path(r"F:\fsy\project\kae_process\infor")
sad_dir = base_dir / "sad"
happy_dir = base_dir / "happy"
out_dir = base_dir / "sad_happy"

os.makedirs(out_dir, exist_ok=True)

files = ["all.txt", "train.txt", "val.txt", "test.txt", "train_val.txt"]

def read_lines_safe(path):
    if not path.exists():
        print(f"缺少文件: {path}")
        return []
    with open(path, "r", encoding="utf-8") as f:
        return f.readlines()

for name in files:
    sad_path = sad_dir / name
    happy_path = happy_dir / name
    out_path = out_dir / name

    sad_lines = read_lines_safe(sad_path)
    happy_lines = read_lines_safe(happy_path)

    # 简单拼接：先 sad 再 happy，如需打乱可再 shuffle
    merged = sad_lines + happy_lines

    with open(out_path, "w", encoding="utf-8") as f:
        f.writelines(merged)

    print(f"合并完成: {out_path}  行数 = {len(merged)}")
```
